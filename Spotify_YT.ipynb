### Top 10 Youtube Artists 
by Views and Likes

```python
df.groupby(['Artist'])['Likes','Views'].sum().reset_index().sort_values(by='Views',ascending=False).head(10)
```

```python

views_df = df.groupby(['Artist'])['Views'].sum().reset_index().sort_values(by='Views',ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.title('Top Artists Based on YouTube Views')
ax = sns.barplot(data =views_df, y='Artist', x='Views', orient='h', palette='Reds')
plt.xlabel('Views in Millions')
plt.ylabel('')
for p in ax.patches:
    value = p.get_width() / 1000000
    ax.text(p.get_width() + 0.1, p.get_y() + 0.5 * p.get_height(), f'{value:.1f}B', ha='left', va='center')

plt.show()
```
